```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
drive.mount("/content/drive/")
data = '/content/drive/Shareddrives/Gain Business Analytics Practicum/Data/Data/decision tree - only23.csv'
```

```python
treedata =  pd.read_csv(data)

treedata.head(20)
```

```python
# Convert categorical variables to numeric
treedata['Region'] = treedata['Region'].astype('category').cat.codes #West 3 South 2 Northeast 1 Midwest 0
treedata['Party'] = treedata['Party'].astype('category').cat.codes #Red 1 Blue 0
```

```python
treedata.head(20)
```

```python
# Decision tree of Party
X = treedata.drop(['Region','Party','State'], axis=1)
y = treedata['Party']

# Build the decision tree
clf = DecisionTreeClassifier(random_state=0,max_depth=3)
clf.fit(X, y)

# Visualize the tree
fig = plt.figure(figsize=(15, 8))
_ = tree.plot_tree(clf, feature_names=X.columns, class_names= ['Blue','Red'], filled=True, fontsize=10)

caption = " Decision Tree based on 'Party'"
plt.figtext(0.5, 0.05, caption, wrap=True, horizontalalignment='center', fontsize=12)

plt.show()
```

```python
# Decision tree of Region
X1 = treedata.drop(['Region','Party','State'], axis=1)
y1 = treedata['Region']

# Build the decision tree
clf = DecisionTreeClassifier(random_state=0,max_depth=3)
clf.fit(X1, y1)

# Visualize the tree
fig = plt.figure(figsize=(15, 8))
_ = tree.plot_tree(clf, feature_names=X.columns, class_names= ["Midwest","Northeast","South","West"], filled=True, fontsize=10)

caption = " Decision Tree based on 'Region'"
plt.figtext(0.5, 0.05, caption, wrap=True, horizontalalignment='center', fontsize=12)

plt.show()
```
